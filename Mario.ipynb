## Detection and logic

### Logic

```python
#1)Running the v8 trained model on an image 
image_path = "Test Image With All Cases.png"
results = modelv8Trained(image_path, classes=[0, 2, 4, 8, 9, 10, 11, 18], iou=0) #classes is used to restrict which class IDs it returns and iou to prevent overlapping

#2)Grid settings 
GRID_ROWS, GRID_COLS = 11, 13
CELL_WIDTH = 640 // GRID_COLS
CELL_HEIGHT = 640 // GRID_ROWS

#3)Class mapping 
classID_name = {
    0: 'big_mario', 1: 'brick_block', 2: 'coin', 3: 'empty_block', 4: 'fire_mario',
    5: 'fireball', 6: 'flower', 7: 'goal_pole', 8: 'goomba', 9: 'hard_block',
    10: 'koopa', 11: 'little_mario', 12: 'mushroom', 13: 'mystery_block', 14: 'pipe',
    15: 'pipe_head', 16: 'shell', 17: 'undestructible_block', 18: 'castle'
}

required_classes = {
    0: 'mario', 4: 'mario', 11: 'mario', # all mario types
    2: 'coin',
    8: 'enemy', 10: 'enemy', #all enemy types
    9: 'block',                           
    18: 'castle'                          
}

#4)Detection to grid mapping
boxes = results[0].boxes.xywh.cpu().numpy()  #bounding box centers and size
classes = results[0].boxes.cls.cpu().numpy()  #class IDs

# a function to convert box center to grid cell
def box_to_grid_cell(x_center, y_center, img_width=640, img_height=640):
    col = int(x_center // (img_width / GRID_COLS))
    row = int(y_center // (img_height / GRID_ROWS))
    return min(max(row, 0), GRID_ROWS - 1), min(max(col, 0), GRID_COLS - 1)

# initializing the grid
grid = [["" for _ in range(GRID_COLS)] for _ in range(GRID_ROWS)]

# grid with detected object labels
for box, cls_id in zip(boxes, classes):
    cls_id = int(cls_id)
    if cls_id in required_classes:
        x_center, y_center, _, height = box
        if cls_id == 18:  # adjusting tge castle detection to the bottom center
            y_center += height / 2
        row, col = box_to_grid_cell(x_center, y_center)
        label = required_classes[cls_id]
        grid[row][col] = label

#5)Drawing the grid and labels on the image to be displayed
annotated_image = results[0].plot()
image = results[0].plot()
annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)
grid_image = annotated_image.copy()

# drawing grid lines on the image
for i in range(1, GRID_ROWS):
    y = i * CELL_HEIGHT
    cv2.line(grid_image, (0, y), (640, y), color=(0, 0, 0), thickness=1)
for j in range(1, GRID_COLS):
    x = j * CELL_WIDTH
    cv2.line(grid_image, (x, 0), (x, 640), color=(0, 0, 0), thickness=1)

# drawing object labels in the grid cells
for r in range(GRID_ROWS):
    for c in range(GRID_COLS):
        label = grid[r][c]
        if label:
            x = c * CELL_WIDTH + 5
            y = r * CELL_HEIGHT + 20
            cv2.putText(grid_image, label, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

# display the grided image with annotations
grid_pil = PIL.Image.fromarray(grid_image)
buf = io.BytesIO()
grid_pil.save(buf, format='PNG')
display(Image(data=buf.getvalue()))

#6)logic planning

DIRECTIONS = {'→': (0, 1), '←': (0, -1), '↑': (-1, 0), '↓': (1, 0), 'A': (0, 1)}

# a function to find all the grid positions for a target label
def find_positions(grid, target):
    return [(r, c) for r in range(GRID_ROWS) for c in range(GRID_COLS) if grid[r][c] == target]

# bfs pathfinding algo
def bfsAlgo(grid, start, targets):
    queue = deque([(start, [])])
    visited = set([start])
    while queue:
        (r, c), path = queue.popleft()
        if (r, c) in targets:
            return path

        for symbol, (dr, dc) in DIRECTIONS.items():
            nr, nc = r + dr, c + dc
            
            if 0 <= nr < GRID_ROWS and 0 <= nc < GRID_COLS and (nr, nc) not in visited:
                cell = grid[nr][nc]
                if cell == 'block':
                    jump_r, jump_c = r - 1, c + 1
                    land_r, land_c = jump_r + 1, jump_c + 2
                    if 0 <= jump_r < GRID_ROWS and 0 <= jump_c < GRID_COLS and grid[jump_r][jump_c] != 'block':
                        queue.append(((jump_r, jump_c), path + ['↑', '→']))
                        visited.add((jump_r, jump_c))
                        queue.append(((land_r, land_c), path + ['↑', '→', '→', '↓']))
                        visited.add((land_r, land_c))
                    continue
                
                elif cell == 'enemy':
                    queue.append(((nr, nc), path + ['A']))
                    visited.add((nr, nc))
                
                elif cell != 'block':
                    queue.append(((nr, nc), path + [symbol]))
                    visited.add((nr, nc))

    return [] #an empty path if no target is found

# Planning the full game path, collecting all the coins and then go to the castle
def gamePlanner(grid):
    grid = [row[:] for row in grid]
    mario_pos = find_positions(grid, 'mario')[0]
    coins = find_positions(grid, 'coin')
    castle = find_positions(grid, 'castle')[0]
    score = 0
    total_path = []
    
    while coins:
        paths = [bfsAlgo(grid, mario_pos, [coin]) for coin in coins]
        best_idx = min(range(len(paths)), key=lambda i: len(paths[i]) if paths[i] else float('inf'))
        path = paths[best_idx]
        if not path:
            break
            
        total_path.extend(path)
        mario_pos = coins[best_idx]
        grid[mario_pos[0]][mario_pos[1]] = ''
        coins.pop(best_idx)
        score += 10
        # Drop down mario after collecting the coin
        drop_r, drop_c = mario_pos[0] + 1, mario_pos[1]
        if 0 <= drop_r < GRID_ROWS and grid[drop_r][drop_c] == '':
            total_path.append('↓')
            mario_pos = (drop_r, drop_c)

    path_to_castle = bfsAlgo(grid, mario_pos, [castle])
    total_path.extend(path_to_castle)

    return total_path, score

#7)Visualizing mario's path frame by frame
def visualize_path(grid, path, start_pos, grid_image):
    r, c = start_pos
    frames = []
    img = grid_image.copy()

    for move in path:
        if move in DIRECTIONS:
            dr, dc = DIRECTIONS[move]
            r, c = r + dr, c + dc

        frame = img.copy()
        top_left = (c * CELL_WIDTH, r * CELL_HEIGHT)
        bottom_right = ((c + 1) * CELL_WIDTH, (r + 1) * CELL_HEIGHT)
        cv2.rectangle(frame, top_left, bottom_right, (0, 0, 255), -1)
        frames.append(frame)

    return frames
```

### Displaying the path

```python
#8)code execution 
action_path, final_score = gamePlanner(grid)
print("Victory Path:", ''.join(action_path))
print("Final Score:", final_score)
```

### Movement visualization

```python
#9)visualization of mario's movement
frames = visualize_path(grid, action_path, find_positions(grid, 'mario')[0], image)
for f in frames:
    plt.imshow(cv2.cvtColor(f, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
